```python
def currency_name(ID):
    response = requests.get('https://www.cbr-xml-daily.ru/daily_json.js')  
    currencies = response.json() 
    vlts = currencies['Valute']
    for v in vlts:
        if vlts[v]['ID'] == ID:
            return vlts[v]['Name']
```

```python
currency_name('R01700J')
```

```python
def wiki_header(url):
    response = requests.get(url)  
    page = BeautifulSoup(response.text, 'html.parser')
    return page.find('h1').text
```

```python
wiki_header('https://en.wikipedia.org/wiki/Operating_system')
```

```python
def get_actors(url):
    response = requests.get(url)  
    page = BeautifulSoup(response.text, 'html.parser')
    
    actors = []
    
    ul_actors = page.find_all('ul', class_ = 'styles_list__I97eu')
    for u in ul_actors:
        a_actors = u.find_all('a')
        for a in a_actors:
            actors.append(a.text)
    return actors
```

```python
get_actors('https://www.kinopoisk.ru/film/42326/')
```

```python
get_actors('https://www.kinopoisk.ru/film/326/')
```

```python
import requests


def get_smm_index(group, token):
    url = 'https://api.vk.com/method/groups.getMembers'

    count = 1000
    offset = 0
    user_ids = []
    max_count = 50000

    params = {
        'group_id': group,
        'v': 5.95,
        'count': count,
        'offset': offset,
        'access_token': token
    }
    # такой же запрос как в прошлый раз
    r = requests.get(url, params=params)
    data = r.json()
    cn = data['response']['count']

    smm = 0
    url = 'https://api.vk.com/method/wall.get'
    params = {
        'domain': group,
        #         'filter': 'all',
        'count': 10,
        'offset': 0,
        'access_token': token,
        'v': 5.95
    }

    response = requests.get(url, params=params)
    for record in response.json()['response']['items'][:]:
        smm += record['comments']['count'] + \
            record['likes']['count'] + record['reposts']['count']
    return smm / cn
```
